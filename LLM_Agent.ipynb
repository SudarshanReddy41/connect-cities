```python
import os
import re
import json
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# === Load LLaMA 3.1 ===
model_name = "meta-llama/Meta-Llama-3-8B-Instruct"  # adjust if using a local model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
llama_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, device_map="auto")

# === Knowledge Base ===
KNOWLEDGE_BASE = {
    "python": "Python is a programming language known for its readability.",
    "openai": "OpenAI is the company that created ChatGPT.",
    "ai": "AI stands for Artificial Intelligence, which simulates human intelligence."
}

# === Utilities ===
def calculate_expression(expression):
    safe_expr = re.sub(r'[^0-9+\-*/(). ]', '', expression)
    if not safe_expr.strip():
        return None
    try:
        return eval(safe_expr)
    except Exception:
        return None

def search_knowledge(query):
    query_lower = query.lower()
    for keyword, value in KNOWLEDGE_BASE.items():
        if keyword in query_lower:
            return value
    return None

# === Tool Selector using LLaMA ===
def select_tool(user_input):
    prompt = (
        f"You are an AI assistant. Based on the following message, "
        f"decide which tool to use: 'calculator', 'knowledge_search', or 'none'. "
        f"Respond ONLY with a JSON object like this:\n"
        f'{{ "tool": "calculator", "input": "5 * (4 + 3)" }}\n\n'
        f"Message: {user_input}"
    )

    try:
        response = llama_pipeline(prompt, max_new_tokens=50, do_sample=False)[0]['generated_text']
        json_start = response.find('{')
        json_end = response.find('}') + 1
        tool_call = json.loads(response[json_start:json_end])
        return tool_call.get("tool"), tool_call.get("input")
    except Exception as e:
        print(f"[Tool Selection Error] {e}")
        return "none", user_input

# === Main Query Handler ===
def handle_query(user_input):
    tool, tool_input = select_tool(user_input)

    if tool == "calculator":
        result = calculate_expression(tool_input)
        return f"The result is: {result}" if result is not None else "I couldn't compute that."

    elif tool == "knowledge_search":
        result = search_knowledge(tool_input)
        return result if result else "I couldn't find relevant information."

    else:
        try:
            response = llama_pipeline(
                f"You are a helpful assistant.\nUser: {user_input}\nAssistant:",
                max_new_tokens=150,
                do_sample=True,
                temperature=0.7
            )
            return response[0]['generated_text'].split("Assistant:")[-1].strip()
        except Exception as e:
            return f"❌ Error: {e}"

# === Main Loop ===
def main():
    print("Welcome! Ask me anything. Type 'quit' to stop.")
    while True:
        user_input = input("You: ")
        if user_input.strip().lower() == "quit":
            print("Agent: Goodbye!")
            break
        response = handle_query(user_input)
        print("Agent:", response)

if __name__ == "__main__":
    main()
```

```python
import os
import re
import json
from llama_cpp import Llama

# === Load LLaMA model ===
llm = Llama(model_path="path/to/llama-3.3-model.gguf")  # adjust to your .gguf file

# === Knowledge Base ===
KNOWLEDGE_BASE = {
    "python": "Python is a programming language known for its readability.",
    "openai": "OpenAI is the company that created ChatGPT.",
    "ai": "AI stands for Artificial Intelligence, which simulates human intelligence."
}

# === Utilities ===
def calculate_expression(expression):
    safe_expr = re.sub(r'[^0-9+\-*/(). ]', '', expression)
    if not safe_expr.strip():
        return None
    try:
        return eval(safe_expr)
    except Exception:
        return None

def search_knowledge(query):
    query_lower = query.lower()
    for keyword, value in KNOWLEDGE_BASE.items():
        if keyword in query_lower:
            return value
    return None

# === LLaMA Tool Selector ===
def select_tool(user_input):
    system_prompt = (
        "You're an AI assistant. Based on the user's message, "
        "decide which tool to use: 'calculator', 'knowledge_search', or 'none'. "
        "Respond ONLY with a JSON object like this:\n"
        "{ \"tool\": \"calculator\", \"input\": \"5 * (4 + 3)\" }"
    )

    prompt = f"{system_prompt}\nUser: {user_input}"
    response = llm(prompt, max_tokens=200, stop=["\n"])
    try:
        tool_call = json.loads(response['choices'][0]['text'].strip())
        return tool_call.get("tool"), tool_call.get("input")
    except Exception as e:
        print(f"[Error selecting tool] {e}")
        return "none", user_input

# === Main Query Handler ===
def handle_query(user_input):
    tool, tool_input = select_tool(user_input)

    if tool == "calculator":
        result = calculate_expression(tool_input)
        return f"The result is: {result}" if result is not None else "I couldn't compute that."

    elif tool == "knowledge_search":
        result = search_knowledge(tool_input)
        return result if result else "I couldn't find relevant information."

    else:
        prompt = f"You are a helpful assistant.\nUser: {user_input}\nAssistant:"
        response = llm(prompt, max_tokens=200)
        return response['choices'][0]['text'].strip()

# === Main Loop ===
def main():
    print("Welcome! Ask me anything. Type 'quit' to stop.")
    while True:
        user_input = input("You: ")
        if user_input.strip().lower() == "quit":
            print("Agent: Goodbye!")
            break
        response = handle_query(user_input)
        print("Agent:", response)

if __name__ == "__main__":
    main()
```
